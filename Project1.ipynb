```python
import pandas as pd


df = pd.read_csv("rat_small.csv") 

# choose a column, we choose latitude here
col = df['Latitude'].dropna()

print("Pandas results:")
print("Mean:", col.mean())
print("Median:", col.median())
print("Mode:", col.mode().iloc[0])
```

```python
# Hard Way

import csv

FILE = "rat_small.csv"
COL = "Latitude"      # column of interest

# turning it into float
values = []
with open(FILE, encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        try:
            values.append(float(row[COL]))
        except:
            pass  

# mean
mean_val = sum(values) / len(values)

# median
sorted_vals = sorted(values)
n = len(sorted_vals)
if n % 2 == 1:
    median_val = sorted_vals[n // 2]
else:
    median_val = (sorted_vals[n//2 - 1] + sorted_vals[n//2]) / 2

# mode
freq = {}
for v in values:
    freq[v] = freq.get(v, 0) + 1

mode_val = max(freq, key=freq.get)

print("Hard way results:")
print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```

```python
# ASCII histogram with fine binwidth = 0.01 degree

bin_width = 0.01

vmin, vmax = min(values), max(values)
start = int(vmin // bin_width) * bin_width
end   = int(vmax // bin_width + 1) * bin_width

# count
from collections import defaultdict
counts = defaultdict(int)

for x in values:
    b = int((x - start) // bin_width)       
    counts[b] += 1

max_count = max(counts.values())
max_bar = 50 if max_count > 0 else 1

print(f"ASCII Histogram for Latitude, bin_width={bin_width:.2f}° (scaled)")
for b in sorted(counts):
    left = start + b * bin_width
    right = left + bin_width
    bar_len = int(counts[b] / max_count * max_bar)
    bar = "*" * bar_len
    print(f"[{left:5.2f}, {right:5.2f}): {bar} ({counts[b]})")
```
